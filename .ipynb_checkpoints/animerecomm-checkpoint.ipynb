```python
import re
def remove_re(text):
    text = re.sub(r'&quot;', '', text)
    text = re.sub(r'.hack//', '', text)
    text = re.sub(r'&#039;', '', text)
    text = re.sub(r'A&#039;s', '', text)
    text = re.sub(r'I&#039;', 'I\'', text)
    text = re.sub(r'&amp;', 'and', text)
    
    return text

anime_data['name'] = anime_data['name'].apply(remove_re)
```

```python
anime_data.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tvf = TfidfVectorizer(min_df = 3,max_features=None,strip_accents = 'unicode',analyzer = 'word',token_pattern=r'\w{1,}',ngram_range=(1,3),stop_words='english')


anime_data['genre'] = anime_data['genre'].fillna('')
genre_str = anime_data['genre'].str.split(',').astype(str)

tvf_matrix = tvf.fit_transform(genre_str)
```

```python
tvf_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

sig = sigmoid_kernel(tvf_matrix, tvf_matrix)
```

```python
indices = pd.Series(anime_data.index, index=anime_data['name']).drop_duplicates()
```

```python
def give_rec(title, sig=sig):
    # Get the index corresponding to original_title
    idx = indices[title]

    # Get the pairwsie similarity scores 
    sig_scores = list(enumerate(sig[idx]))

    # Sort the movies 
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 10 most similar movies
    sig_scores = sig_scores[1:11]

    # Movie indices
    anime_indices = [i[0] for i in sig_scores]

    # Top 10 most similar movies
    return pd.DataFrame({'Anime name': anime_data['name'].iloc[anime_indices].values,
                                 'Rating': anime_data['rating'].iloc[anime_indices].values})
```
